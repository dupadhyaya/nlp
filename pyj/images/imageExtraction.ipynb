```python
import cv2
#import facedetect
import glob
import os
import matplotlib.pyplot as plt
```

```python
dir1 = '/Users/du/downloads/groupPhotos'
dir2 = dir1+/
```

```python
os.listdir(dir1)
```

```python
img = cv2.imread(dir1 + '/image5.jpeg')
```

```python
img[1:2]
```

```python
grayImg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
edges = cv2.Canny(grayImg1, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show();
```

```python
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(grayImg1, None)
image_with_sift = cv2.drawKeypoints(img, keypoints, None)
plt.imshow(cv2.cvtColor(image_with_sift, cv2.COLOR_BGR2RGB))
plt.title('SIFT Features')
plt.show();
```
